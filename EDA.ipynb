```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from dateutil import parser
```

## Data Acquisition

```python
# Data file locations
CUSTOMER_DATA = "data/customer_data.csv"
TOP_SCREENS_DATA = "data/top_screens.csv"
```

```python
# Load Customer Data
dataset = pd.read_csv(CUSTOMER_DATA)
dataset.columns
```

```python
# Load Top Screens Data
top_screens = pd.read_csv(TOP_SCREENS_DATA).top_screens.values
top_screens
```

## Data Cleaning

```python
# Convert hour coloumn from string to int
dataset['hour'] = dataset.hour.str.slice(1, 3).astype(int)
```

## Data Plotting

```python
essential_data = dataset.copy().drop(columns = ['user', 'screen_list', 'enrolled_date', 'first_open', 'enrolled'])
essential_data.head()
```

### Histograms

```python
plt.suptitle('Histograms of Numerical Columns', fontsize=20)
for i in range(1, essential_data.shape[1] + 1):
    plt.subplot(3, 3, i)
    f = plt.gca()
    f.set_title(essential_data.columns.values[i - 1])

    vals = np.size(essential_data.iloc[:, i - 1].unique())
    
    plt.hist(essential_data.iloc[:, i -1], bins=vals, color='#3F5D7D')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

### Correlation with Response Variable

```python
# Find correlation with "enrolled" variable
essential_data.corrwith(dataset.enrolled).plot.bar(figsize=(20,10),
                        title = 'Correlation with Reposnse variable',
                        fontsize = 15, rot = 45, grid = True)
```

## Correlation Matrix

```python
sn.set(style="white", font_scale=2)

# Compute the correlation matrix
corr = essential_data.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(18, 15))
f.suptitle("Correlation Matrix", fontsize = 40)

# Generate a custom diverging colormap
cmap = sn.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sn.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

## Feature Engineering

```python
# Formatting Date Columns
dataset.dtypes
dataset["first_open"] = [parser.parse(row_date) for row_date in dataset["first_open"]]
dataset["enrolled_date"] = [parser.parse(row_date) if isinstance(row_date, str) else row_date for row_date in dataset["enrolled_date"]]
dataset.dtypes
```

```python
# Selecting Time For Response
dataset["difference"] = (dataset.enrolled_date-dataset.first_open).astype('timedelta64[h]')
response_hist = plt.hist(dataset["difference"].dropna(), color='#3F5D7D')
plt.title('Distribution of Time-Since-Screen-Reached')
plt.show()
```

```python
plt.hist(dataset["difference"].dropna(), color='#3F5D7D', range = [0, 100])
plt.title('Distribution of Time-Since-Screen-Reached (First 100 Hours)')
plt.show()
```

```python
dataset.loc[dataset.difference > 48, 'enrolled'] = 0
dataset = dataset.drop(columns=['enrolled_date', 'difference', 'first_open'])
```

### Formatting the screen_list Field

```python
# Mapping Screens to Fields
dataset["screen_list"] = dataset.screen_list.astype(str) + ','

for sc in top_screens:
    dataset[sc] = dataset.screen_list.str.contains(sc).astype(int)
    dataset['screen_list'] = dataset.screen_list.str.replace(sc + ",", "")
```

```python
# Calculate left-over screen
dataset['Other'] = dataset.screen_list.str.count(",")
dataset = dataset.drop(columns=['screen_list'])
```

### Drop correlated fields

```python
# Funnel for saving
savings_screens = ["Saving1",
                    "Saving2",
                    "Saving2Amount",
                    "Saving4",
                    "Saving5",
                    "Saving6",
                    "Saving7",
                    "Saving8",
                    "Saving9",
                    "Saving10"]

dataset["SavingCount"] = dataset[savings_screens].sum(axis=1)
dataset = dataset.drop(columns=savings_screens)
```

```python
# Funnel for credit
cm_screens = ["Credit1",
               "Credit2",
               "Credit3",
               "Credit3Container",
               "Credit3Dashboard"]

dataset["CMCount"] = dataset[cm_screens].sum(axis=1)
dataset = dataset.drop(columns=cm_screens)
```

```python
# Funnel for category
cc_screens = ["CC1",
              "CC1Category",
              "CC3"]

dataset["CCCount"] = dataset[cc_screens].sum(axis=1)
dataset = dataset.drop(columns=cc_screens)
```

```python
# Funnel for loan
loan_screens = ["Loan",
                "Loan2",
                "Loan3",
                "Loan4"]

dataset["LoansCount"] = dataset[loan_screens].sum(axis=1)
dataset = dataset.drop(columns=loan_screens)
```
